```python
import os
from dotenv import load_dotenv
import requests

# .env 파일 로드
load_dotenv('key.env')

# 환경 변수에서 API 키 가져오기
API_KEY= os.getenv("ecos_key")

# 한국은행 API 키 (본인의 API 키로 변경 필요)


# 한국은행 환율 API URL
BASE_URL = "https://ecos.bok.or.kr/api/StatisticSearch/{format}/{api_key}/json/kr/1/1000/731Y001/M/{start_date}/{end_date}/{currency}"

# 데이터 요청 함수
def get_exchange_rate(start_date, end_date, currency_code):
    url = BASE_URL.format(
        format="json", api_key=API_KEY, start_date=start_date, end_date=end_date, currency=currency_code
    )
    response = requests.get(url)
    data = response.json()
    
    # 데이터 변환
    df = pd.DataFrame(data["StatisticSearch"]["row"])
    df["DATE"] = pd.to_datetime(df["TIME"], format="%Y%m")  # 날짜 변환
    df["EXCHANGE_RATE"] = df["DATA_VALUE"].astype(float)  # 환율 데이터 변환
    
    return df[["DATE", "EXCHANGE_RATE"]]

# 예제: USD/KRW 환율 데이터 가져오기 (2022~2024년)
usd_exchange_rate = get_exchange_rate("202201", "202412", "0001")  # 0001 = USD/KRW
print(usd_exchange_rate.head())
```

```python
import requests
import os
from dotenv import load_dotenv

# .env 파일 로드
load_dotenv("key.env")

# 환경 변수에서 API 키 가져오기
API_KEY = os.getenv("ecr_key")

# ExchangeRate-API 엔드포인트 (API 키는 반드시 본인 것으로 변경)
url = f"https://v6.exchangerate-api.com/v6/{API_KEY}/latest/USD"

# API 요청
response = requests.get(url)
data = response.json()

# 환율 데이터 확인
if data["result"] == "success":
    usd_to_krw = data["conversion_rates"]["KRW"]
    print(f"현재 USD → KRW 환율: {usd_to_krw}")
else:
    print("환율 데이터를 가져오지 못했습니다.")

```
