```python
import pandas as pd
df=pd.read_csv('data/df_top15.csv')
```

```python
# 봄철(3, 4, 5월) 데이터 필터링
spring_months = [3, 4, 5]
df_spring = df[df['월'].isin(spring_months)]

# 국적별 봄철 평균 입국자 수 계산
df_spring_avg = df_spring.groupby('국적지역')['입국자수'].mean().reset_index()

# 봄철 평균과 연평균을 병합하여 비교
df_compare_spring = pd.merge(df_spring_avg, df_avg, on='국적지역')
df_compare_spring['봄철 증가율'] = (df_compare_spring['입국자수'] - df_compare_spring['연평균입국자수']) / df_compare_spring['연평균입국자수'] * 100

# 증가율이 높은 (즉, 봄에 입국자가 많은) 나라 찾기
df_spring_favored = df_compare_spring.sort_values(by='증가율', ascending=False)

# 결과 출력
df_spring_favored.head(5)
```

```python
# 여름철(6, 7, 8월) 데이터 필터링
summer_months = [6, 7, 8]
df_summer = df[df['월'].isin(summer_months)]

# 국적별 여름철 평균 입국자 수 계산
df_summer_avg = df_summer.groupby('국적지역')['입국자수'].mean().reset_index()

# 여름철 평균과 연평균을 병합하여 비교
df_compare_summer = pd.merge(df_summer_avg, df_avg, on='국적지역')
df_compare_summer['여름철 증가율'] = (df_compare_summer['입국자수'] - df_compare_summer['연평균입국자수']) / df_compare_summer['연평균입국자수'] * 100

# 증가율이 높은 (즉, 여름에 입국자가 많은) 나라 찾기
df_summer_favored = df_compare_summer.sort_values(by='증가율', ascending=False)

df_summer_favored.head(5)
```

```python
# 가을철(9, 10, 11월) 데이터 필터링
autumn_months = [9, 10, 11]
df_autumn = df[df['월'].isin(autumn_months)]

# 국적별 가을철 평균 입국자 수 계산
df_autumn_avg = df_autumn.groupby('국적지역')['입국자수'].mean().reset_index()

# 가을철 평균과 연평균을 병합하여 비교
df_compare_autumn = pd.merge(df_autumn_avg, df_avg, on='국적지역')
df_compare_autumn['가을철 증가율'] = (df_compare_autumn['입국자수'] - df_compare_autumn['연평균입국자수']) / df_compare_autumn['연평균입국자수'] * 100

# 증가율이 높은 (즉, 가을에 입국자가 많은) 나라 찾기
df_autumn_favored = df_compare_autumn.sort_values(by='증가율', ascending=False)

df_autumn_favored.head(5)
```

```python
# 겨울철(11, 12, 1월) 데이터 필터링
winter_months = [11, 12, 1]
df_winter = df[df['월'].isin(winter_months)]
# 겨울철(11, 12, 1월) 데이터 필터링
df_winter_avg = df_winter.groupby('국적지역')['입국자수'].mean().reset_index()

# 겨울철 평균과 연평균을 병합하여 비교
df_compare_winter = pd.merge(df_winter_avg, df_avg, on='국적지역')
df_compare_winter['겨울철 증가율'] = (df_compare_winter['입국자수'] - df_compare_winter['연평균입국자수']) / df_compare_winter['연평균입국자수'] * 100

# 증가율이 높은 (즉, 겨울에 입국자가 많은) 나라 찾기
df_winter_favored = df_compare_winter.sort_values(by='증가율', ascending=False)

# 결과 출력
df_winter_favored.head(5)
```

```python
# 계절별 데이터 필터링 및 증가율 계산 함수
def calculate_seasonal_growth(df, season_months, season_name):
    df_season = df[df['월'].isin(season_months)]
    df_season_avg = df_season.groupby('국적지역')['입국자수'].mean().reset_index()
    df_compare_season = pd.merge(df_season_avg, df_avg, on='국적지역')
    df_compare_season[f'{season_name} 증가율'] = (df_compare_season['입국자수'] - df_compare_season['연평균입국자수']) / df_compare_season['연평균입국자수'] * 100
    return df_compare_season[['국적지역', f'{season_name} 증가율']]

# 계절별 증가율 계산
df_spring_growth = calculate_seasonal_growth(df, [3, 4, 5], "봄철")
df_summer_growth = calculate_seasonal_growth(df, [6, 7, 8], "여름철")
df_autumn_growth = calculate_seasonal_growth(df, [9, 10, 11], "가을철")
df_winter_growth = calculate_seasonal_growth(df, [11, 12, 1], "겨울철")

# 모든 계절별 증가율을 하나의 데이터프레임으로 병합
df_seasonal_growth = df_avg.copy()
df_seasonal_growth = df_seasonal_growth.merge(df_spring_growth, on='국적지역', how='left')
df_seasonal_growth = df_seasonal_growth.merge(df_summer_growth, on='국적지역', how='left')
df_seasonal_growth = df_seasonal_growth.merge(df_autumn_growth, on='국적지역', how='left')
df_seasonal_growth = df_seasonal_growth.merge(df_winter_growth, on='국적지역', how='left')

# 결과 확인
df_seasonal_growth
```

```python
#df_seasonal_growth의 봄철 증가율이 가장 높은 5개 국가
df_seasonal_growth.sort_values(by='봄철 증가율', ascending=False).head(5)
```

```python
#df_seasonal_growth 저장
df_seasonal_growth.to_csv('data/df_seasonal_growth.csv', index=False)
```
