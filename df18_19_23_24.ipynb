#18 19 23 24 를 합친다.

```python
import pandas as pd
```

```python
df18=pd.read_csv('data/18df.csv')
```

```python
df19=pd.read_csv('data/19df.csv')
```

```python
df23_24=pd.read_csv('data/23_24df.csv')
```

```python
df18
```

```python
df19
```

```python
df23_24
```

```python
# 18, 19, 23_24 데이터프레임을 합친다.
df=pd.concat([df18,df19,df23_24]).reset_index(drop=True)
```

```python
df[df['국적지역']=='중국']
```

```python
#국적지역별 방문자수 합계
sumdf=df.groupby(['국적지역'])['입국자수'].sum().reset_index().sort_values(by='입국자수',ascending=False)
```

```python
sumdf
```

```python
#국가별 총 입국자수 상위 15개국만 추출
sumdf.head(15)
```

```python
#해당 국가만 리스트 추출
countrylist=sumdf.head(15)['국적지역'].tolist()
```

```python
countrylist
```

```python
#df 에서 countrylist 에 들어있는 나라만 추출
df_top15=df[df['국적지역'].isin(countrylist)]
```

```python
df_top15
```

```python
#타임시리즈 예측을위해 년,월 컬럼을 날짜형식으로 변환
df_top15['날짜'] = pd.to_datetime(df_top15['년'].astype(str) + '-' + df_top15['월'].astype(str).str.zfill(2))

# 국가별 데이터 정렬
df_top15=df_top15.sort_values(by=['국적지역', '날짜']).reset_index(drop=True)
```

```python
# 저장
df_top15.to_csv('data/df_top15.csv',index=False)
```
